```python
import pandas as pd
EDA = pd.read_csv('data/raw/adult.csv')
EDA.head()
```

```python
EDA.isna().sum()
```

```python
Nan = EDA.workclass.map(lambda desc: "?" in desc).sum()
print(Nan)
```

```python
import numpy as np
EDA.replace('?',np.nan,inplace=True)
EDA.isna().sum()
```

```python
EDA.dropna(inplace=True)
EDA.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
EDA.education = le.fit_transform(EDA.education)
```

```python
print("Category Mapping:", le.classes_)
```

```python
EDA.income = le.fit_transform(EDA.income)
EDA.relationship = le.fit_transform(EDA.relationship)
EDA.race = le.fit_transform(EDA.race)
EDA.sex = le.fit_transform(EDA.sex)
```

```python
EDA.head()
```

```python
EDA = pd.get_dummies(EDA, columns=['workclass','marital.status','occupation'])
```

```python
EDA.head()
```

```python
top_n = 10  # оставляем топ-10 стран
target_col = 'native.country'

# 1️⃣ Находим топ-N самых частых категорий
top_categories = EDA[target_col].value_counts().nlargest(top_n).index
# 2️⃣ Заменяем все остальные на "Other"
EDA[target_col] = EDA[target_col].apply(lambda x: x if x in top_categories else 'Other')
EDA = pd.get_dummies(EDA, target_col)
```

```python
EDA.head()
```

```python
y = EDA.income
X = EDA.drop('income',axis=1)
```

```python
X.head()
```

```python
y.head()
```

```python
DF_ABT= X.copy()
DF_ABT['income']= y
DF_ABT.to_csv('data/processed/abt_sample.csv', index=False)
```
